```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from matplotlib import cm
import math
```

```python
f = xr.open_dataset("All_2014_NCEP_1stprcesd_gridded.nc").load()
f.close()
f = f.where(f.CTRC != -999.)
```

```python
myfontsize = 12

fig, axs = plt.subplots(1,1,figsize=[6/2.54, 6/2.54],
                        sharex = True, sharey = True,
                        dpi = 200)
var = (f.TMP_500mb.values.flatten()/(f.CTT.values.flatten() + 273.15))**4.
weights = np.ones_like(var)/float(len(var))
axs.hist(var, 40,weights=weights)
axs.set_xlabel('$ (T_{500}/T_c)^{4}$')
axs.set_ylabel('Probability density')

plt.show()
fig.savefig('FigS2_Ta_Tc_histogram.png', dpi=fig.dpi, bbox_inches='tight')
```
